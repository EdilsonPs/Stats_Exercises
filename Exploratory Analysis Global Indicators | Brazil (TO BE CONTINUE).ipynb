```python
import pandas as pd
pd.set_option('display.max_columns', 82)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/bquxjob_29f94aa4_1875336cc5e.csv")
```

```python
df.head(3)
```

```python
df_brazil = df.loc[df.country == "Brazil"]
```

```python
df_brazil.head(3)
```

```python
def plot(column):
  sns.set_context("poster")
  plt.figure(figsize=(20,1))
  sns.lineplot(data = df_brazil, x = df_brazil["year"], y = df[column])
  plt.ylabel(" ")
  plt.title(column)
  plt.show()
```

```python
plot("social_progress_index")
```
